```python
# 1.Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
sorted_df_DH1=df["DH1"].sort_values()
print(sorted_df_DH1)
```

```python
#Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
sorted_df_DH2= df.sort_values(by=["DH2","GT"],ascending=[True,True])
print(sorted_df_DH2)
```

```python
#Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median,
def Q1(g):
  return np.percentile(g,25)
def Q2(g):
  return np.percentile(g,55)
def Q3(g):
  return np.percentile(g,75)
def std(g):
  return np.std(g)
```

```python
# Thống kế dữ liệu dạng group
groupKT_DH1= df.groupby(["KT"])["DH1"].agg(["count","sum", "mean", "median", "min", "max", std, Q1, Q2, Q3])
groupKT_DH1
```

```python
#Có thể thay thế cho dạng bảng bằng 
##df.pivot_table()
```

```python
#Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, sdt, Q1, Q2 và Q3 của DH1 theo KT và KV
groupKT_KV_DH1= df.groupby(["KT","KV"])["DH1"].agg(["count","sum", "mean", "median", "min", "max", std, Q1, Q2, Q3])
groupKT_KV_DH1
```

```python
#Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, sdt, Q1, Q2 và Q3 của DH1 theo KT, KV và DT
groupKT_KV_DT_DH1= df.groupby(["KT","KV","DT"])["DH1"].agg(["count","sum", "mean", "median", "min", "max", std, Q1, Q2, Q3])
groupKT_KV_DT_DH1
```

```python
# Thống kê số lượng theo từng khu vực
StaKV = df.groupby("KV")["KV"].agg(["count"])
StaKV.plot.pie(
    autopct="%.2f",
    fontsize=20,
    subplots=True,
    figsize=(8, 4))
```

```python
StaKT = df.groupby("KT")["KT"].agg(["count"])
StaKT.plot.pie(
    autopct="%.2f",
    fontsize=20,
    subplots=True,
    figsize=(8, 4))
```

```python
#Hãy trình bày dữ liệu biến KV với các học sinh là nam thuộc dân tộc Kinh, có điểm thõa mãn điều kiện (DH1 >= 5.0 và DH2 >= 4.0 và DH3 >= 4.0)
StaGT = df.groupby("GT")["GT"].agg(["count"])
StaGT.plot.bar()
```
