# **Phần 1:**

Thống kê dữ liệu

1. Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
2. Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
3. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, sdt, Q1, Q2 và Q3 của DH1 theo KT
4. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, sdt, Q1, Q2 và Q3 của DH1 theo KT và KV
5. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, sdt, Q1, Q2 và Q3 của DH1 theo KT, KV và DT

```python
# 1.Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
# Ý tưởng sử dụng hàm "df.sort_values()" để sắp xếp theo thứ tự tăng dần
dulieuxettuyendaihoc_sort_DH1=dulieuxettuyendaihoc.sort_values(by='DH1')
print(dulieuxettuyendaihoc_sort_DH1)
```

```python
#2.Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
# Sử dụng hàm "df.sort_values()"
dulieuxettuyendaihoc_sort_DH2_GT= dulieuxettuyendaihoc.sort_values(by=["DH2","GT"])
print(dulieuxettuyendaihoc)
```

```python
#3. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, sdt, Q1, Q2 và Q3 của DH1 theo KT
#Ý tưởng tạo 4 hàm Q1, Q2, Q3 và sdt
#dùng np.percentile để xét từng khoản điểm (https://numpy.org/doc/stable/reference/generated/numpy.percentile.html)
def Q1(g):
  return np.percentile(g,25)
def Q2(g):
  return np.percentile(g,50)
def Q3(g):
  return np.percentile(g,75)
def sdt(g):
  return np.sdt(g)

# sử dụng module pivot trong pandas để thống kê các giá trị count, sum, mean, median, min, max, sdt...
#Tham khảo:(https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html)
pivot_KT_table_DH1 = pd.pivot_table(dulieuxettuyendaihoc, values='DH1', index=['KT'],aggfunc=['count','sum','median','min','max','std',Q1,Q2,Q3])
pivot_KT_table_DH1
```

```python
#4.Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, sdt, Q1, Q2 và Q3 của DH1 theo KT và KV
pivot_KT_KV_table_DH1=pd.pivot_table(dulieuxettuyendaihoc, values="DH1", index=['KT','KV'], aggfunc=['count','sum','median','min','max','std',Q1,Q2,Q3])
pivot_KT_KV_table_DH1
```

```python
#5.Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, sdt, Q1, Q2 và Q3 của DH1 theo KT, KV và DT
pivot_KT_KV_DT_table_DH1=pd.pivot_table(dulieuxettuyendaihoc, values="DH1", index=['KT','KV','DT'], aggfunc=['count','sum','median','min','max','std',Q1,Q2,Q3])
pivot_KT_KV_DT_table_DH1
```
